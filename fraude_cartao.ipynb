```python
data1 = data.sample(frac = 0.1,random_state=1)
data1.shape
```

```python
Fraude = data1[data1['Class']==1]
Valido = data1[data1['Class']==0]
outlier_fraction = len(Fraude)/float(len(Valido))
print(outlier_fraction)
```

```python
print("Quantidade de Fraudes : {}".format(len(Fraude)))
print("Transações Validas: {}".format(len(Valido)))
```

```python
corrmat = data1.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,25))
g=sns.heatmap(data[top_corr_features].corr(),annot=True,cmap="coolwarm")
plt.savefig("matrixsns.png")
```

```python
columns = data1.columns.tolist()
columns = [c for c in columns if c not in ["Class"]]
target = "Class"
state = np.random.RandomState(42)
X = data1[columns]
Y = data1[target]
X_outliers = state.uniform(low=0, high=1, size=(X.shape[0], X.shape[1]))
print(X.shape)
print(Y.shape)
```

```python
# Model Prediction 
```

```python
classifiers = {
    "Isolation Forest":IsolationForest(n_estimators=100, max_samples=len(X), contamination=outlier_fraction, random_state=state,
                                       verbose=0),
    "Local Outlier Factor":LocalOutlierFactor(n_neighbors=20, algorithm='auto', leaf_size=30, metric='minkowski', p=2,
                                              metric_params=None, contamination=outlier_fraction),
    "Suport Vector Machine":OneClassSVM(kernel='rbf', degree=3, gamma=0.1, nu=0.05, max_iter=-1, random_state=state),
         
}
```

```python
type(classifiers)
```

```python
n_outliers = len(Fraude)
for i, (clf_name,clf) in enumerate(classifiers.items()):
    if clf_name == 'Local Outlier Factor':
        y_pred = clf.fit_predict(X)
        scores_prediction = clf.negative_outlier_factor_
    elif clf_name == 'Support Vector Machine':
        clf.fit(X)
        y_pred = clf.predict(X)
    else:
        clf.fit(X)
        scores_prediction = clf.decision_function(X)
        y_pred = clf.predict(X)
        
    y_pred[y_pred == 1] = 0
    y_pred[y_pred == -1] = 1
    n_errors = (y_pred != Y).sum()
    
    print("{}: {}".format(clf_name,n_errors))
    print("Precisão(%) :")
    print(accuracy_score(Y,y_pred))
    print("Classificação :")
    print(classification_report(Y,y_pred))
```
